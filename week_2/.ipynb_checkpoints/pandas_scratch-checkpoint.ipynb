```python
#selects value count of the column!
drinks['continent'].value_counts()
```

```python
#What's missing? North America! Why? Because it reads as 'NA' in the data frame#
#slice missing values

mask = drinks['continent'].isnull()
drinks[mask]
```

```python
drinks[mask]['continent']='NA'
```

```python
#we haven't actually assigned this to the data frame, this is just a view
#would need to assign to the original data frame
drinks.loc[drinks['continent'].isnull()].fillna('NA')
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
#question: what day of week has the highest tips?
#split-apply-combine method

tips.groupby('day')
```

```python
tips.groupby('day').mean()
```

```python
tips.groupby('day')['tip'].mean()
```

```python
tips.groupby(['day','smoker'])['tip'].mean()
```

```python
tips.groupby(['time'])['tip'].mean()
```

```python
tips.groupby(['sex']).mean()
```

```python
tips['tip_pct'] = tips['tip']/tips['total_bill']
```

```python
tips.head()
```

```python
tips.groupby('sex')['tip_pct'].mean()
```

```python
tips.groupby(['sex'])['tip_pct'].mean().plot(kind = 'bar')
```
